**Group by in DataFrame**

```python
df1.groupby('sex').agg({'age': 'count'})
# this code groups the values of the column â€˜sexâ€™ and aggregates the column â€˜ageâ€™ by the count of that column.
```

```python
df1.groupby('sex').agg({'age':'mean'}) # this code groups the values of the column â€˜sexâ€™ and aggregates the column â€˜ageâ€™ by the mean of that column.
```

```python
df1.groupby(['cp', 'sex']).agg({'age': 'count'}) #Group multiple columns
```
